```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

def process_stock_data(file_path, seq_length=60):
    data = pd.read_csv(file_path)
    data['Date'] = pd.to_datetime(data['Date'])
    data = data.sort_values('Date')

    features = ['High', 'Low', 'Open', 'Volume']
    target = ['Close']
    X_data = data[features].values
    y_data = data[target].values

    scaler_X = MinMaxScaler(feature_range=(0, 1))
    scaler_y = MinMaxScaler(feature_range=(0, 1))
    X_scaled = scaler_X.fit_transform(X_data)
    y_scaled = scaler_y.fit_transform(y_data)

    def create_sequences(X_data, y_data, seq_length):
        X = []
        y = []
        for i in range(seq_length, len(X_data)):
            X.append(X_data[i-seq_length:i, :])
            y.append(y_data[i, 0])
        return np.array(X), np.array(y)

    X, y = create_sequences(X_scaled, y_scaled, seq_length)

    train_size = int(len(X) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dense(units=25))
    model.add(Dense(units=1))
    model.compile(optimizer='adam', loss='mean_squared_error')

    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.001)
    model.fit(X_train, y_train, validation_split=0.2, batch_size=64, epochs=20, callbacks=[early_stopping, reduce_lr])

    predictions = model.predict(X_test)
    predictions = scaler_y.inverse_transform(predictions)
    y_test_actual = scaler_y.inverse_transform(y_test.reshape(-1, 1))

    rmse = np.sqrt(mean_squared_error(y_test_actual, predictions))
    mape = mean_absolute_percentage_error(y_test_actual, predictions) * 100
    mbe = np.mean(y_test_actual - predictions)

    results = pd.DataFrame({
        'Date': data['Date'][-len(y_test):].values,
        'Actual_Close': y_test_actual.flatten(),
        'Predicted_Close': predictions.flatten()
    })

    return results, rmse, mape, mbe

stock_files = [
    'Goldman_Sachs_stock_data.csv',
    'Johnson and Johnson_stock_data.csv',
    'JP Morgan Chase and Co_stock_data.csv',
    'Nike_stock_data.csv',
    'Pfizer_stock_data.csv'
]

stocks = [
    'Goldman Sachs',
    'JNJ',
    'JP Morgan',
    'Nike',
    'Pfizer'
]

method = '[LSTM]'

for index, file_path in enumerate(stock_files):
    print(f"Processing {file_path}...")
    results, rmse, mape, mbe = process_stock_data(file_path)
    
    output_file = file_path.replace('.csv', '_predictions_LSTM.csv')
    results.to_csv(output_file, index=False)
    
    print(f"Results saved to {output_file}")
    print(f"RMSE: {rmse}")
    print(f"MAPE: {mape:.2f}%")
    print(f"MBE: {mbe}")
    print("-" * 50)

    plt.figure(figsize=(10, 6))
    plt.plot(results['Date'], results['Actual_Close'], 's-', label='Actual Closing Price', color='blue', linewidth=1, markersize=3)
    plt.plot(results['Date'], results['Predicted_Close'], 'o-', label='Predicted Closing Price', color='red', linewidth=1, markersize=3)

    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.title(f'ORiginal vs Predicted Closing Prices')
    plt.legend()
    
    num_ticks = 4  
    step = max(1, len(results['Date']) // num_ticks)
    plt.xticks(results['Date'][::step])
    
    plt.text(0.95, 0.05, f'{stocks[index]} {method}',
             horizontalalignment='right',
             verticalalignment='bottom',
             transform=plt.gca().transAxes,
             fontsize=14)

    plt.legend(fontsize=12, loc='upper left')
    plt.tight_layout()
    plt.savefig(f'./pictures/{stocks[index]}_LSTM.png', bbox_inches='tight')
    plt.show()
```
